### Librerias

```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('flight_delays.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# se cambia actualizan las columnas desde la 5 a la 7, desde object a time

# Lista de las columnas a convertir
cols_to_fix = [
    'ScheduledDeparture',
    'ActualDeparture',
    'ScheduledArrival',
    'ActualArrival'
]

# Convertir a datetime
for col in cols_to_fix:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# Verificar el cambio
print(df.info())
```

```python
# Se elimina la columna
df = df.drop(columns=['FlightID'])
df.head()
```

```python
df.describe().round(2)
```

```python
df.isna().sum()
```

```python
#se revisa, el motivo de los retrazos, cantidad de estos sobre el total, y se cruza cantida de minutos. se valida que los valores nan, corresponden vuelos sin retrazos.

# Llenamos los valores nulos para que aparezcan en el anlisis
df['DelayReason'] = df['DelayReason'].fillna('Sin retraso reportado')

# Agrupamos por motivo
resumen = df.groupby('DelayReason').agg(
    Cantidad=('DelayMinutes', 'count'),
    Suma_Minutos=('DelayMinutes', 'sum'),
    Promedio_Minutos=('DelayMinutes', 'mean')
).reset_index()

# Calculamos el porcentaje sobre el total de registros
total_vuelos = resumen['Cantidad'].sum()
resumen['Porcentaje (%)'] = (resumen['Cantidad'] / total_vuelos * 100).round(2)

# Reordenamos las columnas para que la tabla sea fcil de leer
resumen = resumen[[
    'DelayReason',
    'Cantidad',
    'Porcentaje (%)',
    'Suma_Minutos',
    'Promedio_Minutos'
]]

# Ordenamos por cantidad para ver lo ms frecuente primero
resumen = resumen.sort_values(by='Cantidad', ascending=False)

# Mostrar el resultado
print(resumen)
```
